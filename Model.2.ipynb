```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

Collect Data + Preprocess Data

```python
tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0][['Symbol', 'GICS Sector']]
tickers['Symbol'] = tickers['Symbol'].str.replace('.', '-')
sector_breakdown = tickers.groupby('GICS Sector')['Symbol'].apply(list)
sector_breakdown = sector_breakdown.to_dict()

# Print the sector breakdown
print(sector_breakdown)
```

```python
ticker_list = []
for sector in sector_breakdown:
    ticker_list.extend(sector_breakdown[sector])

today = pd.Timestamp.today().strftime('%Y-%m-%d')
month_ago = pd.Timestamp.today() - pd.DateOffset(months=5)

data = yf.download(ticker_list, start=month_ago, end=today)
#data = data.drop(columns=['Open', 'High', 'Low', 'Close', 'Volume'])
data
```

```python
# Backward fill NaN values
data = data.fillna(method='bfill', axis=0)

# Backward fill NaN values
data = data.fillna(method='ffill', axis=0)

# Print number of NA values across each column
print(data.isna().sum(axis=1))
```

```python
# Calculate daily returns
returns = data.pct_change()

# Drop the index name Adj Close
returns = returns.droplevel(0, axis=1)

# Group tickers in returns dataframe by sector using sector_breakdown's keys & values
sector_mean_returns = {}

# Extract each key & value pair from sector_breakdown
for sector, tickers in sector_breakdown.items():
    # Get tickers for each sector
    sector_data = returns[tickers]

    # Calculate mean returns for each sector and add to dictionary
    sector_mean_returns[sector] = sector_data.mean(axis=1)

# Convert sector_mean_returns dict to a DataFrame
sector_mean_returns = pd.DataFrame(sector_mean_returns)

sector_mean_returns
```

```python
# Drop rows that have NaN values
sector_mean_returns = sector_mean_returns.dropna()
sector_mean_returns
```

```python
# Calculate the variance-covariance matrix for the 11 sector returns
cov_matrix = sector_mean_returns.cov()
cov_matrix
```

Black-Litterman Model

```python
class BlackLittermanModel:
    def __init__(self, market_equilibrium_returns, covariance_matrix, views, confidence_matrix):
        self.market_equilibrium_returns = market_equilibrium_returns
        self.covariance_matrix = covariance_matrix
        self.views = views
        self.confidence_matrix = confidence_matrix

    def calculate_bl_expected_returns(self):
        """
        Calculate expected returns using the Black-Litterman model.
        """
        tau = 0.025  # Tau is the scaling factor, adjust as needed

        # Calculate the equilibrium excess returns
        pi = np.dot(self.covariance_matrix, self.market_equilibrium_returns)

        # Calculate the Black-Litterman expected returns
        bl_expected_returns = np.dot(
            np.linalg.inv(np.linalg.inv(tau * self.covariance_matrix) + np.dot(np.dot(self.views.T, np.linalg.inv(self.confidence_matrix)), self.views)),
            np.dot(np.linalg.inv(tau * self.covariance_matrix), pi) + np.dot(np.dot(self.views.T, np.linalg.inv(self.confidence_matrix)), self.views.T.dot(self.views.dot(self.market_equilibrium_returns)))
        )

        return bl_expected_returns
```

```python
# Assuming you have a risk-free rate and equity risk premium
risk_free_rate = 0.047  # Adjust based on the current risk-free rate
equity_risk_premium = 0.06  # Adjust based on economic indicators

# Calculate market equilibrium returns
market_equilibrium_returns = risk_free_rate + equity_risk_premium + sector_mean_returns.mean(axis=0)

# Print the market equilibrium returns
print("Market Equilibrium Returns:")
print(market_equilibrium_returns)
```

```python
# Define the sectors in your portfolio
sectors = [
    'Communication Services',
    'Consumer Discretionary',
    'Consumer Staples',
    'Energy',
    'Financials',
    'Health Care',
    'Industrials',
    'Information Technology',
    'Materials',
    'Real Estate',
    'Utilities'
]

# Assuming you have len(sectors) assets in your portfolio
n_assets = len(sectors)

# Define your views based on your analysis
views = np.zeros(n_assets)

views[0] = 0.025  # Positive view on Communication Services
views[1] = -0.015  # Negative view on Consumer Discretionary
views[2] = 0.02   # Positive view on Consumer Staples
views[3] = -0.015  # Negative view on Energy
views[4] = 0.02  # Higher positive view on Financials
views[5] = 0.015  # Higher slightly positive view on Health Care
views[6] = -0.20  # Negative view on Industrials
views[7] = 0.04  # Higher positive view on Information Technology
views[8] = -0.012  # Negative view on Materials
views[9] = 0.01  # Higher slightly positive view on Real Estate
views[10] = -0.018  # Negative view on Utilities

# Adjust the values to make them add up to 1
total_views = sum(views)
views = [view / total_views for view in views]
total_views = sum(views)
print(total_views)

print("Your Views:")
print(views)

# Ensure that the sum of views is zero for market equilibrium
views = views - np.mean(views)
print("Adjusted Views:")
print(views)

# Define confidence levels for each view
confidence_levels = {
    'Communication Services': 0.09,
    'Consumer Discretionary': 0.09,
    'Consumer Staples': 0.09,
    'Energy': 0.09,
    'Financials': 0.09,
    'Health Care': 0.09,
    'Industrials': 0.09,
    'Information Technology': 0.09,
    'Materials': 0.09,
    'Real Estate': 0.09,
    'Utilities': 0.09,
}

# Create a diagonal matrix using the confidence levels
confidence_matrix = np.diag([confidence_levels[sectors[i]] for i in range(n_assets)])
```

```python
# Assuming you have market_equilibrium_returns, covariance_matrix, investor_views, and confidence_matrix defined
bl_model = BlackLittermanModel(market_equilibrium_returns, cov_matrix, views, confidence_matrix)
bl_expected_returns = bl_model.calculate_bl_expected_returns()

# Print the Black-Litterman expected returns
print("Black-Litterman Expected Returns:")
print(bl_expected_returns)
```

LSTM Model

```python
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
def create_sequences(data, sequence_length):
    """
    Create input sequences for LSTM model.

    Parameters:
    - data: numpy array, input data (scaled returns)
    - sequence_length: int, length of each sequence

    Returns:
    - X: numpy array, input sequences
    - y: numpy array, corresponding output values
    """
    X, y = [], []

    for i in range(len(data) - sequence_length):
        sequence = data[i:i + sequence_length]
        target = data[i + sequence_length]
        X.append(sequence)
        y.append(target)

    return np.array(X), np.array(y)
```

```python
# Assuming you have already calculated mean_returns
equilibrium_expected_returns = risk_free_rate + equity_risk_premium + sector_mean_returns.mean(axis=0)

# Print Equilibrium Expected Returns
print("Equilibrium Expected Returns:")
print(equilibrium_expected_returns)

# Assuming you have already loaded the data and calculated returns
cov_matrix = sector_mean_returns.cov()

# Print Covariance Matrix
print("Covariance Matrix:")
print(cov_matrix)

# LSTM-based Portfolio Optimization

# Prepare data for LSTM
returns_data = sector_mean_returns[sectors].copy()
scaler_lstm = MinMaxScaler(feature_range=(0, 1))
scaled_returns_data = scaler_lstm.fit_transform(returns_data)

# Create input sequences for LSTM
sequence_length_lstm = 10
sequences_lstm, targets_lstm = create_sequences(scaled_returns_data, sequence_length_lstm)

# Split data into train and test sets
train_size_lstm = int(len(sequences_lstm) * 0.8)
train_lstm, test_lstm = sequences_lstm[:train_size_lstm], sequences_lstm[train_size_lstm:]
train_targets, test_targets = targets_lstm[:train_size_lstm], targets_lstm[train_size_lstm:]

# Extract features (X) and target variable (y) for LSTM
X_train_lstm, y_train_lstm = train_lstm[:, :-1, :], train_targets
X_test_lstm, y_test_lstm = test_lstm[:, :-1, :], test_targets

# Reshape input data for LSTM (batch_size, time_steps, features)
X_train_lstm = np.reshape(X_train_lstm, (X_train_lstm.shape[0], X_train_lstm.shape[1], X_train_lstm.shape[2]))
X_test_lstm = np.reshape(X_test_lstm, (X_test_lstm.shape[0], X_test_lstm.shape[1], X_test_lstm.shape[2]))

print("X_train_lstm shape:", X_train_lstm.shape)
print("y_train_lstm shape:", y_train_lstm.shape)

```

```python
# Build a separate LSTM model for each sector
models_lstm = []
predictions_lstm = []

for i in range(len(sectors)):
    # Build LSTM model
    model_lstm = Sequential()
    model_lstm.add(LSTM(units=50, return_sequences=True, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
    model_lstm.add(LSTM(units=50))
    model_lstm.add(Dense(units=1))
    model_lstm.compile(optimizer='adam', loss='mean_squared_error')

    # Train the LSTM model
    model_lstm.fit(X_train_lstm, y_train_lstm[:, i], epochs=10, batch_size=32)

    # Make predictions on the test set
    predictions_sector = model_lstm.predict(X_test_lstm)
    predictions_lstm.append(predictions_sector)

    # Save the model for each sector
    models_lstm.append(model_lstm)
```

```python
# Convert predictions list to numpy array
predictions_lstm = np.array(predictions_lstm)

# Reshape predictions back to the original shape
predictions_lstm = np.reshape(predictions_lstm, (predictions_lstm.shape[1], predictions_lstm.shape[0]))

# Inverse transform the predictions using the scaler
predicted_returns_lstm = scaler_lstm.inverse_transform(predictions_lstm)

# Reshape back to the original shape
predicted_returns_lstm = np.reshape(predicted_returns_lstm, (predicted_returns_lstm.shape[0], len(sectors)))

# Print the predicted returns for each sector
for i, sector in enumerate(sectors):
    print(f"Predicted Returns for {sector}:")
    print(predicted_returns_lstm[:, i])
```

```python
# Initialize an empty dictionary to store combined expected returns for each sector
combined_expected_returns_by_sector = {}
weight_lstm = 0.2

# Iterate through each sector
for sector in sectors:
    # Equilibrium expected returns for the current sector
    equilibrium_expected_returns_sector = equilibrium_expected_returns[sector]

    try:
        # LSTM predictions for the current sector
        predicted_returns_lstm_sector = predicted_returns_lstm[sectors.index(sector)]
    except IndexError:
        print(f"Skipping {sector} because it doesn't have LSTM predictions.")
        continue

    # Combine expected returns for the current sector
    combined_expected_returns_sector = equilibrium_expected_returns_sector + weight_lstm * predicted_returns_lstm_sector

    # Store the combined expected returns for the current sector
    combined_expected_returns_by_sector[sector] = combined_expected_returns_sector

# Print the combined expected returns for each sector
for sector, combined_returns in combined_expected_returns_by_sector.items():
    print(f"Combined Expected Returns for {sector}:")
    print(combined_returns)
    print()
```

Combine Black-Litterman and LSTM

```python
# Assuming you have already calculated mean_returns for the LSTM predictions
equilibrium_expected_returns_lstm = risk_free_rate + equity_risk_premium + predicted_returns_lstm.mean(axis=0)

# Assuming you have already loaded the data and calculated returns for the LSTM predictions
cov_matrix_lstm = np.cov(predicted_returns_lstm, rowvar=False)

# Align LSTM predictions with sectors
aligned_predicted_returns_lstm = np.zeros(len(sectors))
for i in range(len(sectors)):
    if i < len(predicted_returns_lstm):
        # Take the average of the predictions for the current sector
        aligned_predicted_returns_lstm[i] = np.mean(predicted_returns_lstm[i])
    else:
        # Handle the case when the sector is not present in LSTM predictions
        print(f"Skipping {sectors[i]} because it doesn't have LSTM predictions.")
        aligned_predicted_returns_lstm[i] = 0  # You can set a default value or handle it differently

# Combine Expected Returns
combined_expected_returns = equilibrium_expected_returns + weight_lstm * aligned_predicted_returns_lstm

# Print Combined Expected Returns
print("Combined Expected Returns:")
print(combined_expected_returns)
```

```python
# Combine Covariance Matrices
combined_cov_matrix = (1 - weight_lstm) * cov_matrix + weight_lstm * cov_matrix_lstm

combined_cov_matrix
```

Calculate Allocations

```python
# Calculate the inverse of the combined covariance matrix
inv_combined_cov_matrix = np.linalg.inv(combined_cov_matrix)

# Calculate the expected excess returns
expected_excess_returns = combined_expected_returns

# Calculate the Black-Litterman expected returns
bl_expected_returns = np.dot(inv_combined_cov_matrix, expected_excess_returns)

# Shift all expected returns to make them positive
bl_expected_returns_shifted = bl_expected_returns - np.min(bl_expected_returns) + 0.001  # Adding a small constant to avoid zeros

# Normalize weights to get allocations
allocations = bl_expected_returns_shifted / np.sum(bl_expected_returns_shifted)

# Print the allocations for each sector
for sector, allocation in zip(sectors, allocations):
    print(f"Allocation for {sector}: {allocation:.4f}")
```

 Evaluate Portfolio Performance

```python
def evaluate_portfolio(returns, weights):
    # Calculate annualized portfolio return
    annual_portfolio_return = np.sum(returns.mean() * weights) * 252

    # Calculate annualized portfolio volatility
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights)))

    risk_free_rate = 0.2

    # Calculate Sharpe ratio
    sharpe_ratio = (annual_portfolio_return - risk_free_rate) / portfolio_volatility

    return annual_portfolio_return, portfolio_volatility, sharpe_ratio

# Assuming you have a DataFrame 'returns_data' containing returns for each sector
# Assuming 'allocations' contains the weights/allocations for each sector

# Evaluate portfolio performance
portfolio_return, portfolio_volatility, sharpe_ratio = evaluate_portfolio(returns_data, allocations)

# Print the results
print(f"Annual Portfolio Return: {portfolio_return:.4f}")
print(f"Annual Portfolio Volatility: {portfolio_volatility:.4f}")
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
```
